```python
import os
import cv2
```

```python
rootpath = './data/videos/'
```

```python
os.listdir(rootpath)
```

```python
for d in os.listdir(rootpath):
    if d != '.DS_Store':
        path = os.path.join(rootpath,d)
        files = os.listdir(path)
        for ind, f in enumerate(files,1):
            p = os.path.join(path, f)
            os.rename(p, os.path.join(path, f+str(ind)))
```

```python
def FrameCapture(path): 
    # Path to video file
    vidObj = cv2.VideoCapture(path)
  
    # Used as counter variable
    count = 0
  
    # checks whether frames were extracted
    success = 1
  
    while success:
  
        # vidObj object calls read
        # function extract frames
        success, image = vidObj.read()
  
        # Saves the frames with frame-count
        cv2.imwrite("frame%d.jpg" % count, image)
  
        count += 1
```

```python
vidcap = cv2.VideoCapture('./data/videos/up/up1.mov')
success,image = vidcap.read()
count = 0
while success:
    try:
        success,image = vidcap.read()
        cv2.imwrite('./data/videos/center/'+"frame%d.jpg" % count, image)     # save frame as JPEG file      
        print('Read a new frame: ', success)
        count += 1
    except:
        pass
```

```python
def dir_to_frames(writepath):
    writepath = writepath
    files = [i for i in os.listdir(writepath) if i != '.DS_Store' and i != '.ipynb_checkpoints']
    count = 1
    for f in files:
        vidpath = os.path.join(writepath, f)
        vidcap = cv2.VideoCapture(vidpath)
        success,image = vidcap.read()
        framecount = 0
        while success:
            try:
                success,image = vidcap.read()
                cv2.imwrite(writepath + str(count)+"frame%d.jpg" % framecount, image)     # save frame as JPEG file      
                print('Read a new frame: ', success)
                framecount += 1
            except:
                pass
        count+=1
```
